```python
l = Counter(df.original_title)
l
```

```python
df = df[df.vote_count.notna()]
```

```python
useful = []
others = []
for i, (film_name, amount) in enumerate(l.items()):
    if amount > 1:
        useful.append(film_name)
    else:
        others.append(list(l.keys()).index(film_name))
```

```python
res = []
for i, film_name in enumerate(useful):
    try:
        d = df[df.original_title == film_name]
        d = d[d.vote_count == d.vote_count.max()]
        res.append(d.index[0])
    except Exception as e:
        print(film_name, e)
        print()
```

```python
res += others
```

```python
result = sorted(res)
result = pd.DataFrame(result)
```
